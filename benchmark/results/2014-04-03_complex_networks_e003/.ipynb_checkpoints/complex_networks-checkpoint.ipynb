```python
import pandas as pd
import pandas.io.sql as pd_sql
import sqlite3 as sql
import matplotlib.pyplot as plt

#get data for kahypar
kahypar_connection = sql.connect("/home/schlag/repo/schlag_git/benchmark/results/2014-04-03_complex_networks_e003/complex_networks.db")
kahypar_data = pd_sql.read_frame("select * from experiments",kahypar_connection)
kahypar_min_cuts = pd.DataFrame(kahypar_data.groupby('graph')['cut'].min()).reset_index()
```

```python
from scipy import stats
print 'KaHyPar: geometric mean (min_cuts)=', stats.gmean(kahypar_min_cuts['cut'])
kahypar_min_cuts
```

```python
kahypar_avg_cuts = pd.DataFrame(kahypar_data.groupby('graph')['cut'].mean()).reset_index()
print 'KaHyPar: geometric mean (avg_cuts)=', stats.gmean(kahypar_avg_cuts['cut'])
kahypar_avg_cuts
```

```python
kahypar_avg_time = pd.DataFrame(kahypar_data.groupby('graph')['time'].mean()).reset_index()
kahypar_avg_time
```
