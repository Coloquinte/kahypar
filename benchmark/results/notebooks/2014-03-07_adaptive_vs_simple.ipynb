```python
import pandas as pd
import pandas.io.sql as pd_sql
import sqlite3 as sql
import matplotlib.pyplot as plt

# get the data for alpha = 10
adaptive_connection = sql.connect("/home/schlag/repo/schlag_git/benchmark/results/2014-03-05_parametertuning_alpha_t100_same_seed_as_i_oldStopRuleWithFix/parametertuning_alpha.db")
alpha10_data = pd_sql.read_frame("select * from experiments where twowayfm_alpha = 10",adaptive_connection)
alpha10_min_cuts = pd.DataFrame(alpha10_data.groupby('graph')['cut'].min()).reset_index()

#get data for simple stopping rule
simple_connection = sql.connect("/home/schlag/repo/schlag_git/benchmark/results/2014-02-26_parametertuning_i/parametertuning_i.db")
simple190_data = pd_sql.read_frame("select * from experiments where twowayfm_fruitless_moves = 190",simple_connection)
simple190_min_cuts = pd.DataFrame(simple190_data.groupby('graph')['cut'].min()).reset_index()

# create dataframe with both min cuts and select the best
both_min_cuts = pd.DataFrame(alpha10_min_cuts)
both_min_cuts = both_min_cuts.rename(columns={'cut' : 'min_cut_adaptive'})
both_min_cuts = pd.merge(both_min_cuts, simple190_min_cuts, on='graph')
both_min_cuts = both_min_cuts.rename(columns={'cut' : 'min_cut_simple'})
both_min_cuts['min'] = both_min_cuts.apply(lambda row: (row['min_cut_adaptive'] 
                                                               if row['min_cut_adaptive'] < row['min_cut_simple']
                                                               else row['min_cut_simple']), axis=1)

#calculate precentage of derivation of mean cuts from min cut for adaptive stopping rule
adaptive_percentages = pd.DataFrame(alpha10_data.groupby('graph')['cut'].mean()).reset_index()
adaptive_percentages = pd.merge(adaptive_percentages, both_min_cuts[['graph','min']], on='graph')
adaptive_percentages = adaptive_percentages.rename(columns={'cut' : 'mean_cut', 'min' : 'min_cut'})
adaptive_percentages['percent_deviation'] = adaptive_percentages.apply(lambda row : ((row['mean_cut'] * 100.0 / row['min_cut'] -100)), axis=1)

#calculate plot data for adaptive stopping rule
adaptive_plot = pd.DataFrame({'deviation_leq' : np.arange(0,16)})
adaptive_plot['num_graphs'] = adaptive_plot.apply(lambda row : (len(adaptive_percentages[adaptive_percentages['percent_deviation']<= row['deviation_leq']].index)),axis=1)
adaptive_plot['percentage_of_graphs'] = adaptive_plot.apply(lambda row : (len(adaptive_percentages[adaptive_percentages['percent_deviation']<= row['deviation_leq']].index) * 100 / len(adaptive_percentages.index)),axis=1)

#calculate precentage of derivation of mean cuts from min cut for simple stopping rule
simple_percentages = pd.DataFrame(simple190_data.groupby('graph')['cut'].mean()).reset_index()
simple_percentages = pd.merge(simple_percentages, both_min_cuts[['graph','min']], on='graph')
simple_percentages = simple_percentages.rename(columns={'cut' : 'mean_cut', 'min' : 'min_cut'})
simple_percentages['percent_deviation'] = simple_percentages.apply(lambda row : ((row['mean_cut'] * 100.0 / row['min_cut'] -100)), axis=1)

#calculate plot data for simple stopping rule
simple_plot = pd.DataFrame({'deviation_leq' : np.arange(0,16)})
simple_plot['num_graphs'] = simple_plot.apply(lambda row : (len(simple_percentages[simple_percentages['percent_deviation']<= row['deviation_leq']].index)),axis=1)
simple_plot['percentage_of_graphs'] = adaptive_plot.apply(lambda row : (len(simple_percentages[simple_percentages['percent_deviation']<= row['deviation_leq']].index) * 100 / len(simple_percentages.index)),axis=1)

#the plot
axis = adaptive_plot.plot(x='deviation_leq', y='percentage_of_graphs', label='adaptive (alpha=10)', title='Deviations of average cuts from best min cut (of both)', style='bx')
axis.set_ylabel('Percentage of Graphs')
simple_plot.plot(ax=axis, x='deviation_leq', y='percentage_of_graphs', label='simple (i=190)', style='r+')
axis.set_xlabel('Deviation from min-cut $[\leq \%]$')
axis.get_xaxis().set_ticks(np.arange(20))
axis.get_yaxis().set_ticks(np.arange(0,105,5))
plt.gcf().set_size_inches(14.5,6.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
