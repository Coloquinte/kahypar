```python
import pandas as pd
import pandas.io.sql as pd_sql
import sqlite3 as sql

#get data for vcycle
vcycle_connection = sql.connect("/home/schlag/repo/schlag_git/benchmark/results/2014-03-07_current_opt_vcycle_correctPQselection/current_opt_vcycle.db")
vcycle_data = pd_sql.read_frame("select * from experiments",vcycle_connection)
vcycle_min_cuts = pd.DataFrame(vcycle_data.groupby('graph')['cut'].min()).reset_index()

#get no vcycle data
normal_connection = sql.connect("/home/schlag/repo/schlag_git/benchmark/results/2014-03-07_current_opt_correct_PQselection/current_opt.db")
normal_data = pd_sql.read_frame("select * from experiments where global_search_iterations = 1",normal_connection)
normal_min_cuts = pd.DataFrame(normal_data.groupby('graph')['cut'].min()).reset_index()

# create dataframe with both min cuts and select the best
both_min_cuts = pd.DataFrame(vcycle_min_cuts)
both_min_cuts = both_min_cuts.rename(columns={'cut' : 'min_cut_vcycle'})
both_min_cuts = pd.merge(both_min_cuts, normal_min_cuts, on='graph')
both_min_cuts = both_min_cuts.rename(columns={'cut' : 'min_cut_normal'})
both_min_cuts['min'] = both_min_cuts.apply(lambda row: (row['min_cut_vcycle'] 
                                                               if row['min_cut_vcycle'] < row['min_cut_normal']
                                                               else row['min_cut_normal']), axis=1)

#calculate precentage of derivation of mean cuts from min cut for kahyper
vcycle_percentages = pd.DataFrame(vcycle_data.groupby('graph')['cut'].mean()).reset_index()
vcycle_percentages = pd.merge(vcycle_percentages, both_min_cuts[['graph','min']], on='graph')
vcycle_percentages = vcycle_percentages.rename(columns={'cut' : 'mean_cut', 'min' : 'min_cut'})
vcycle_percentages['percent_deviation'] = vcycle_percentages.apply(lambda row : ((row['mean_cut'] * 100.0 / row['min_cut'] -100)), axis=1)

#calculate plot data for adaptive stopping rule
vycle_plot = pd.DataFrame({'deviation_leq' : np.arange(0,27)})
vycle_plot['num_graphs'] = vycle_plot.apply(lambda row : (len(vcycle_percentages[vcycle_percentages['percent_deviation']<= row['deviation_leq']].index)),axis=1)
vycle_plot['percentage_of_graphs'] = vycle_plot.apply(lambda row : (len(vcycle_percentages[vcycle_percentages['percent_deviation']<= row['deviation_leq']].index) * 100 / len(vcycle_percentages.index)),axis=1)

#calculate precentage of derivation of mean cuts from min cut for hmetis
normal_percentages = pd.DataFrame(normal_data.groupby('graph')['cut'].mean()).reset_index()
normal_percentages = pd.merge(normal_percentages, both_min_cuts[['graph','min']], on='graph')
normal_percentages = normal_percentages.rename(columns={'cut' : 'mean_cut', 'min' : 'min_cut'})
normal_percentages['percent_deviation'] = normal_percentages.apply(lambda row : ((row['mean_cut'] * 100.0 / row['min_cut'] -100)), axis=1)

#calculate plot data for simple stopping rule
normal_plot = pd.DataFrame({'deviation_leq' : np.arange(0,27)})
normal_plot['num_graphs'] = normal_plot.apply(lambda row : (len(normal_percentages[normal_percentages['percent_deviation']<= row['deviation_leq']].index)),axis=1)
normal_plot['percentage_of_graphs'] = normal_plot.apply(lambda row : (len(normal_percentages[normal_percentages['percent_deviation']<= row['deviation_leq']].index) * 100 / len(normal_percentages.index)),axis=1)

# the plot
axis = vycle_plot.plot(x='deviation_leq', y='percentage_of_graphs', label='KaHyPar (10 Vcycles)', title='Deviations of average cuts from best min cut (of both)', style='bx')
axis.set_ylabel('Percentage of Graphs')
normal_plot.plot(ax=axis, x='deviation_leq', y='percentage_of_graphs', label='KaHyPar (no Vcycle)', style='r+')
axis.set_xlabel('Deviation from min-cut $[\leq \%]$')
axis.get_xaxis().set_ticks(np.arange(30))
axis.get_yaxis().set_ticks(np.arange(0,105,5))
plt.gcf().set_size_inches(14.5,6.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
