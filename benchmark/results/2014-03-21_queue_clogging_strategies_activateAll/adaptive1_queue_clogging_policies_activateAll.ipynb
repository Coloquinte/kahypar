```python
import pandas as pd
import pandas.io.sql as pd_sql
import sqlite3 as sql
import matplotlib.pyplot as plt

connection = sql.connect("/home/schlag/repo/schlag_git/benchmark/results/2014-03-21_queue_clogging_strategies_activateAll/adaptive1_queue_clogging.db")
doNotRemoveAny = pd_sql.read_frame("select * from experiments where QueueCloggingPolicy = 'DoNotRemoveAnyCloggingEntriesAndResetEligiblity'", connection)
onlyRemoveIfBothClogged = pd_sql.read_frame("select * from experiments where QueueCloggingPolicy = 'OnlyRemoveIfBothQueuesClogged'", connection)
removeOnlyTheCloggedOne = pd_sql.read_frame("select * from experiments where QueueCloggingPolicy = 'RemoveOnlyTheCloggingEntry'", connection)

doNotRemoveAny_minCuts = pd.DataFrame(doNotRemoveAny.groupby('graph')['cut'].min()).reset_index()
onlyRemoveIfBothClogged_minCuts = pd.DataFrame(onlyRemoveIfBothClogged.groupby('graph')['cut'].min()).reset_index()
removeOnlyTheCloggedOne_minCuts = pd.DataFrame(removeOnlyTheCloggedOne.groupby('graph')['cut'].min()).reset_index()
minCuts = doNotRemoveAny_minCuts
minCuts = minCuts.rename(columns = {'cut' : 'doNotRemoveAny'})
minCuts = pd.merge(minCuts, onlyRemoveIfBothClogged_minCuts, on='graph')
minCuts = minCuts.rename(columns = {'cut' : 'onlyRemoveIfBothClogged'})
minCuts = pd.merge(minCuts, removeOnlyTheCloggedOne_minCuts, on='graph')
minCuts = minCuts.rename(columns = {'cut' : 'removeOnlyTheCloggedOne'})
minCuts
```

```python
from scipy import stats
print 'doNotRemoveAny gmean(min_cuts)=', stats.gmean(minCuts['doNotRemoveAny'])
print 'onlyRemoveIfBothClogged gmean(min_cuts)=', stats.gmean(minCuts['onlyRemoveIfBothClogged'])
print 'removeOnlyTheCloggedOne gmean(min_cuts)=', stats.gmean(minCuts['removeOnlyTheCloggedOne'])
```

```python
doNotRemoveAny_avg = pd.DataFrame(doNotRemoveAny.groupby('graph')['cut'].mean()).reset_index()
onlyRemoveIfBothClogged_avg = pd.DataFrame(onlyRemoveIfBothClogged.groupby('graph')['cut'].mean()).reset_index()
removeOnlyTheCloggedOne_avg = pd.DataFrame(removeOnlyTheCloggedOne.groupby('graph')['cut'].mean()).reset_index()
avg_cuts = doNotRemoveAny_avg
avg_cuts = avg_cuts.rename(columns = {'cut' : 'doNotRemoveAny'})
avg_cuts = pd.merge(avg_cuts, onlyRemoveIfBothClogged_avg, on='graph')
avg_cuts = avg_cuts.rename(columns = {'cut' : 'onlyRemoveIfBothClogged'})
avg_cuts = pd.merge(avg_cuts, removeOnlyTheCloggedOne_avg, on='graph')
avg_cuts = avg_cuts.rename(columns = {'cut' : 'removeOnlyTheCloggedOne'})
avg_cuts
```

```python
print 'doNotRemoveAny gmean(avg_cuts)=', stats.gmean(avg_cuts['doNotRemoveAny'])
print 'onlyRemoveIfBothClogged gmean(avg_cuts)=', stats.gmean(avg_cuts['onlyRemoveIfBothClogged'])
print 'removeOnlyTheCloggedOne gmean(avg_cuts)=', stats.gmean(avg_cuts['removeOnlyTheCloggedOne'])
```
